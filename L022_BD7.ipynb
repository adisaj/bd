```python
# AMS algorithm
insert_no = int(input('Enter the number of elements to be inserted: '))
insert_list = []

for i in range(insert_no):
    insert_list.append(int(input('Enter element: ')))

print('Inserted elements:', insert_list)

# Indices to select elements
x1 = 1
x2 = 3
x3 = 5

# Selected elements based on 1-based index
x1new = insert_list[x1 - 1]
x2new = insert_list[x2 - 1]
x3new = insert_list[x3 - 1]

print('x1:', x1new)

# Function to count occurrences including the initial position
def repeated(x, arr, start_index):
    count = 0
    # Count the occurrences at the start_index
    if arr[start_index] == x:
        count += 1
    # Now count occurrences after the start_index
    for i in range(start_index + 1, len(arr)):
        if x == arr[i]:
            count += 1
    print(f"Total occurrences of {x} from index {start_index}: Count: {count}")
    return count

# Count occurrences of selected elements
r1 = repeated(x1new, insert_list, x1 - 1)  # Counting occurrences of x1new
r2 = repeated(x2new, insert_list, x2 - 1)  # Counting occurrences of x2new
r3 = repeated(x3new, insert_list, x3 - 1)  # Counting occurrences of x3new

print('Counts:', r1, r2, r3)

# AMS function definition
def ams_fun(n, r):
    return (2 * r - 1) * n

# Calculate AMS values
f1 = ams_fun(insert_no, r1)
f2 = ams_fun(insert_no, r2)
f3 = ams_fun(insert_no, r3)

print('AMS Values:', f1, f2, f3)

# Calculate surprise number
surprise_no = (f1 + f2 + f3) / 3
print('Surprise Number:', surprise_no)
```
